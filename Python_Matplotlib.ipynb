**3. Write a Python program to plot two or more lines on same plot with suitable legends of each line. Go to the editor **

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = []
y2 = []

for i in x1:
    y1.append(i*2)
    y2.append(i*4)

plt.plot(x1, y1, 'bs', label='Multiplication of 2', linewidth=10.0)
plt.plot(x1, y2, 'r', label='Multiplication of 4', linewidth=5.0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Multiplication")
plt.grid()
plt.legend()
plt.show()
    
```

**4. Plot multiple lines graph with setproperies method.**

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = []
y2 = []

for i in x1:
    y1.append(i*2)
    y2.append(i*4)

lines = plt.plot(x1, y1, x1, y2)
plt.setp(lines[0], color='b', linestyle='solid', label='M2')
plt.setp(lines[1], color='r', linestyle='--', linewidth='5.0', label='M4')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Multiplication")
plt.grid()
plt.legend()
plt.show()
```

**5. Multiple plots**

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = []
y2 = []

for i in x1:
    y1.append(i*2)
    y2.append(i*4)

plt.figure(1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Multiplication")

plt.subplot(211)
plt.plot(x1, y1)

plt.subplot(212)
plt.plot(x1, y2)


plt.grid()
plt.show()
```
